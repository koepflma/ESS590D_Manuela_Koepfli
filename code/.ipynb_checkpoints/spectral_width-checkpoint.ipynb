### Load Packages

````python
import covseisnet as csn
import numpy as np
import pandas as pd
import obspy
import matplotlib.pyplot as plt
from obspy import UTCDateTime
import time
from obspy.clients.fdsn import Client
import warnings
warnings.simplefilter('ignore')
````

### List of Seismic Stations

````python
client = Client("IRIS")  

s1  = 'UW-EDM-EHZ'
s2  = 'UW-SHW-EHZ'
s3  = 'UW-HSR-EHZ'
s4  = 'UW-SOS-EHZ'
s5  = 'UW-JUN-EHZ'
s6  = 'UW-ELK-EHZ'
s7  = 'UW-TDL-EHZ'
s8  = 'UW-SUG-EHZ'
s9  = 'UW-YEL-EHZ'
s10 = 'UW-FL2-EHZ'
s11 = 'UW-CDF-?H?'

s12 = 'UW-SEP-?H?'
s13 = 'CC-SEP-?H?'
# s14 = 'UW-STD-EHZ'
s15 = 'CC-STD-BHZ'

s16 = 'CC-VALT-BH?'
s17 = 'CC-JRO-BHZ'
s18 = 'CC-HOA-BH?'
s19 = 'CC-LOO-BH?'
s20 = 'CC-USFR-BH?'
s21 = 'CC-NED-EHZ'
s22 = 'CC-REM-BH?'
s23 = 'CC-SWFL-BH?'
s24 = 'CC-SFW2-BH?'
s25 = 'CC-MIDE-EHZ'
s26 = 'CC-MIBL-EHZ'
s27 = 'CC-BLIS-EHZ'
s28 = 'CC-RAFT-EHZ'
s29 = 'CC-SPN5-EHZ'
s30 = 'CC-SEND-EHZ'

list_stations = [s1,s2,s3,s4,s5,s6,s7,s8,s9,s10,s11,s12,s13,
                 s15,s16,s17,s18,s19,s20,s21,s22,s23,s24,s25,s26,s27,s28,s29,s30]
````

### Set time of interesst

````python
year = 2004
jday = 268
hour = 0 # start hour
signal_duration_sec = 1 * 3600 # hour * sec data you are loading
````

### Set Parameters for Spectral Width

````python
window_duration_sec = 10 # time window in sec
average = 10 # in time windows
````

### Load Seismic Data for all available Stations

````python
stime = time.time() # start time
stream = csn.arraystream.ArrayStream() # create an emtpty array stream

t = UTCDateTime(year=year, julday=jday, hour=hour) # input time and date to UTCDateTime
for xx in list_stations: # loop over station list
    netstacha = xx.split('-')
    net = netstacha[0] # network
    sta = netstacha[1] # station
    cha = netstacha[2] # channel
    try: # try to load the data and the XLM file
        st = client.get_waveforms(net, sta, '', cha, t, t + signal_duration_sec) # read seismic data

        st.merge(fill_value=0) # mearge traces and fill with 0 if there is a gap between traces

        # correct insrument response
        inv = obspy.read_inventory('/auto/pnwstore1-wd11/PNWStationXML/{}/{}.{}.xml'.format(net,net,sta)) # load station XML
        pre_filt = [1e-3, 5e-2, 45, 50] # set bandpass filter frequencies
        water_level = 60

        st[0].remove_response(inventory=inv, zero_mean=True,taper=True, taper_fraction=0.05,
                                  pre_filt=pre_filt, output="VEL", water_level=water_level,
                                  plot=False) # remove instrument response --> output velocity

        # correct positive dip
        dip = inv.get_orientation(st[0].id, datetime=st[0].stats.starttime)['dip']
        if dip > 0:
            st[0].data *= -1

        st.detrend('linear') # detrend each trace
        st.taper(max_percentage=0.05, type='hann') # taper each trace
            
        # downsample data to 25 Hz
        st.resample(25)
            
        stream.append(st[0]) # append to strem (created before loop)

    except:
        print(netstacha, jday)
        pass

print('Loading data tooks {} seconds.'.format(round(time.time()-stime),3))
    
# synchronize data
raw_stream = stream.synchronize(start=t, duration_sec=signal_duration_sec, method="linear") 
raw_stream
````

### Plot the spectral widthout preprocessing

````python
stime = time.time() # start time
stream = raw_stream.copy() # copy used for different pre-processing

# create plot for preprocessed data with one-bit spectral whitening
fig, ax = plt.subplots()

# calculate covariance from stream
times, frequencies, covariances = csn.covariancematrix.calculate(
    stream, window_duration_sec, average# , window_step_sec=window_step_sec
)

# calculate spectral width
spectral_width = covariances.coherence(kind="spectral_width")

# show network covariance matrix spectral width
#fig, ax = plt.subplots(1, constrained_layout=True)
img = ax.pcolormesh(
    hour + times / 3600, frequencies, spectral_width.T, rasterized=True, cmap="inferno_r",
    vmin=0, vmax=1)

ax.set_ylim([0, stream[0].stats.sampling_rate / 2])
ax.set_xlabel("{} (hours)".format(t.date))
ax.set_ylabel("Frequency (Hz)")
ax.set_title("")
plt.colorbar(img, ax=ax).set_label("Covariance matrix spectral width")

print('Calculation tooks {} seconds.'.format(round(time.time()-stime),3))
````

### Plot the spectral width of different pre-processed steps

````python
stime = time.time() # start time
stream = raw_stream.copy() # copy used for different pre-processing

# create plot #1 - no preprocessin------------------------------------------------------------------
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(15, 10))

# calculate covariance from stream
times, frequencies, covariances1 = csn.covariancematrix.calculate(
    stream, window_duration_sec, average
)


# calculate spectral width
spectral_width1 = covariances1.coherence(kind="spectral_width")

# show network covariance matrix spectral width
#fig, ax = plt.subplots(1, constrained_layout=True)
img = ax1.pcolormesh(
    hour + times / 3600, frequencies, spectral_width1.T, rasterized=True, cmap="inferno_r",
    vmin=0, vmax=1.)

ax1.set_ylim([0, stream[0].stats.sampling_rate / 2])
ax1.set_xlabel("{} (hours)".format(t.date))
ax1.set_ylabel("Frequency (Hz)")
ax1.set_title("No preprocessing")
plt.colorbar(img, ax=ax1).set_label("Covariance matrix spectral width")
print('First plot tooks {} seconds.'.format(round(time.time()-stime),3))

# create plot #2 - preprocess data with one-bit spectral whitening----------------------------------
stime2 = time.time() # start time
stream = raw_stream.copy()

# preprocess using default setting of one-bit spectral whitening
stream.preprocess()

# calculate covariance from stream
times, frequencies, covariances2 = csn.covariancematrix.calculate(
    stream, window_duration_sec, average
)

# calculate spectral width
spectral_width2 = covariances2.coherence(kind="spectral_width")

# show network covariance matrix spectral width
#fig, ax = plt.subplots(1, constrained_layout=True)
img = ax2.pcolormesh(
    hour + times / 3600, frequencies, spectral_width2.T, rasterized=True, cmap="inferno_r",
    vmin=0.2, vmax=1.8)

ax2.set_ylim([0, stream[0].stats.sampling_rate / 2])
ax2.set_xlabel("{} (hours)".format(t.date))
ax2.set_ylabel("Frequency (Hz)")
ax2.set_title("One-bit spectral whitening")
plt.colorbar(img, ax=ax2).set_label("Covariance matrix spectral width")
print('Second plot tooks {} seconds.'.format(round(time.time()-stime2),3))

# create plot #3 - preprocess data with one-bit temporal normalization------------------------------
stime3 = time.time() # start time
stream = raw_stream.copy()

# preprocess using one-bit temporal normalization
stream.preprocess(domain="temporal", method="onebit")

# calculate covariance from stream
times, frequencies, covariances3 = csn.covariancematrix.calculate(
    stream, window_duration_sec, average
)

# calculate spectral width
spectral_width3 = covariances3.coherence(kind="spectral_width")

# show network covariance matrix spectral width
#fig, ax = plt.subplots(1, constrained_layout=True)
img = ax3.pcolormesh(
    hour + times / 3600, frequencies, spectral_width3.T, rasterized=True, cmap="inferno_r",
    vmin=0.2, vmax=1.8)

ax3.set_ylim([0, stream[0].stats.sampling_rate / 2])
ax3.set_xlabel("{} (hours)".format(t.date))
ax3.set_ylabel("Frequency (Hz)")
ax3.set_title("One-bit temporal normalization")
plt.colorbar(img, ax=ax3).set_label("Covariance matrix spectral width")
print('Third plot tooks {} seconds.'.format(round(time.time()-stime3),3))

# create plot #4 - preprocess data with smooth spectral whitening and temporal normalization--------
stime4 = time.time() # start time
stream = raw_stream.copy()

# preprocess using smooth spectral whitening and temporal normalization
stream.preprocess(domain="spectral", method="smooth")
stream.preprocess(domain="temporal", method="smooth")

# calculate covariance from stream
times, frequencies, covariances4 = csn.covariancematrix.calculate(
    stream, window_duration_sec, average
)

# calculate spectral width
spectral_width4 = covariances4.coherence(kind="spectral_width")

# show network covariance matrix spectral width
#fig, ax = plt.subplots(1, constrained_layout=True)
img = ax4.pcolormesh(
    hour + times / 3600, frequencies, spectral_width4.T, rasterized=True, cmap="inferno_r",
    vmin=0.2, vmax=1.8)

ax4.set_ylim([0, stream[0].stats.sampling_rate / 2])
ax4.set_xlabel("{} (hours)".format(t.date))
ax4.set_ylabel("Frequency (Hz)")
ax4.set_title("Smooth spectral and temporal preprocessing")
plt.colorbar(img,ax=ax4).set_label("Covariance matrix spectral width")
print('Fourth plot tooks {} seconds.'.format(round(time.time()-stime4),3))
print('Total time is {} seconds.'.format(round(time.time()-stime),3))
plt.savefig('../plots/{}_{}_{}_dt{}_wd{}_av{}.png'.format(year,jday,hour,signal_duration_sec, int(window_duration_sec), average), dpi=300, bbox_inches='tight')
````

````python
feature_list = ['UTC', 'Lat [Â°]', 'Lon [Â°]', 'depth [m]', 'event_type', 'mag', 'event']
df = pd.DataFrame(0, index=np.arange(len(events)), columns=feature_list)

for ii in range (0, len(events)):        
    df['UTC'].loc[ii] = events[ii].origins[0].time.datetime
    df['Lat [Â°]'].loc[ii] = events[ii].origins[0].latitude
    df['Lon [Â°]'].loc[ii] = events[ii].origins[0].longitude
    df['depth [m]'].loc[ii] = events[ii].origins[0].depth    
    df['event_type'].loc[ii] = events[ii].event_type   
    df['mag'].loc[ii] = events[ii].magnitudes[0].mag    
    df['event'].loc[ii] = 1
    
df.loc[ii+1] = [t1.datetime, np.nan, np.nan, np.nan, np.nan, np.nan,0] # add starttime
df.loc[ii+2] = [t2.datetime, np.nan, np.nan, np.nan, np.nan, np.nan,0] # add endtime
df = df.set_index('UTC')
print(df.shape)
df
````
